```python
df = pd.read_csv(r"/content/drive/MyDrive/kulyah/Datasets/dataset_twitter/dataset_tweet_sentimen_tayangan_tv.csv")
df.head(10)
```

# **Case Folding**

```python
# menghapus row kososng
df['Text Tweet'].dropna(inplace=True)
```

```python
# lowercase
df['Text Tweet'] = [entry.lower() for entry in df['Text Tweet']]
```

```python
# menghapus angka, karakter, http, hashtag, @username
df['Text Tweet'] = [' '.join(re.sub("([@#][A-Za-z0-9\w+:\/\/\S+]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|(\d+)"," ",tweet).split()) for tweet in df['Text Tweet']]
```

```python
df.head(10)
```

# **Filtering (Stopwords Removal), Stemming**

```python
# stopwords removal
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

# stemming
factory = StemmerFactory()
stemmer = factory.create_stemmer()

tweet_bersih = []

for tweet in df['Text Tweet']:
  stop = stopword.remove(tweet) # stopwords removal process
  stem = stemmer.stem(stop) # stemming process
  tweet_bersih.append(stem)

df['tweet_bersih'] = tweet_bersih
print(df['tweet_bersih'])
```
